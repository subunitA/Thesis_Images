```python
import seaborn #ONLY works with linear regression
#while looking for help in documentation, whereever you see an *sns*, replace in your head with
#seaborn; I think they added that just to type less

import matplotlib

df = pandas.read_csv('pnp_Std.csv')
ax = seaborn.regplot(df.mM, df.Abs, color='goldenrod')
ax.set_xlabel('p-nitrophenol Concentration (mM)')
ax.set_ylabel('Absorbance at 405 nm')
ax.set_xlim(left=-0.05)
ax.set_ylim(bottom=-0.1)
ax.set_title('Absorbance of pNP  Standards')

ax.figure.savefig('pnp_std.png')


```
